## Checking dataset for null values

Different ways to check for a null value in the dataset

```python
# Check for Null Values in the Entire DataFrame
null_values_red = df_red.isna()
null_values_white = df_white.isna()
print(null_values_red)
print(null_values_white)
```

```python
# Count Null Values in Each Column
null_counts_red = df_red.isna().sum()
null_counts_white = df_white.isna().sum()
print(null_counts_red)
print("-------------")
print(null_counts_white)
```

```python
# Count Null Values in the Entire DataFrame
total_nulls_red = df_white.isna().sum().sum()
total_nulls_white = df_red.isna().sum().sum()
print(total_nulls_red)
print(total_nulls_white)
```

```python
# Percentage of Null Values in Each Column
null_percentage_red = df_red.isna().mean() * 100
null_percentage_white = df_white.isna().mean() * 100
print(null_percentage_red)
print("-------------")
print(null_percentage_white)
```

This dataset, was previously cleaned by its creator, however, checking for null values is the first step of data wrangling/cleaning and part of the data analyst/scientist job.

## Binning process

 Binning is when you group values together

### Binning Red Wine

```python
# Define bins and labels
# let the computer decide what was classified as low, medium and high
bins_defined_function = np.linspace(min(df_red["alcohol"]), max(df_red["alcohol"]), 4) # defined by numpy function
bins_defined = [0,12.5,13.5,30]  #source: https://www.wineinvestment.com/learn/magazine/2019/06/wines-alcohol-levels-explained/
group_names = ["Low", "Medium", "High"]
```

```python
# Bin the data
df_red["alcohol-binned"] = pd.cut(df_red["alcohol"], bins_defined_function, labels=group_names, include_lowest = True)
```

```python
# Plot histogram
plt.figure(figsize=(10, 6))
df_red["alcohol-binned"].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Alcohol Content Categories')
plt.ylabel('Quantity')
plt.title('Histogram of Binned Alcohol Content')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bin the data
df_red["alcohol-binned"] = pd.cut(df_red["alcohol"], bins_defined, labels=group_names, include_lowest = True)
```

```python
# Plot histogram
plt.figure(figsize=(10, 6))
df_red["alcohol-binned"].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Alcohol Content Categories')
plt.ylabel('Quantity')
plt.title('Histogram of Binned Alcohol Content')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

As you can see, letting the function define the bins for you gives a difference in what you consider low, medium, and high. It all depends on the type of analysis that you are doing and what are your end goals, but if you want to follow strict procedures and measures it is always good to ask and check real world measurements.

### Types of correlation:
<b>Positive Correlation:</b> As one variable increases, the other also increases. For example, height and weight typically have a positive correlation.

<b>Negative Correlation:</b> As one variable increases, the other decreases. An example is the relationship between the speed of a car and the time taken to reach a destination.

<b>No Correlation:</b> There is no discernible relationship between the variables. For instance, the correlation between the amount of coffee consumed and intelligence.

<b>Perfect Correlation:</b>

Perfect Positive Correlation (r = +1): The variables move together perfectly.

Perfect Negative Correlation (r = -1): One variable moves perfectly in the opposite direction to the other.

<b>Linear Correlation:</b> The relationship between the variables can be represented by a straight line. This is what is typically meant by "correlation" in statistics.
